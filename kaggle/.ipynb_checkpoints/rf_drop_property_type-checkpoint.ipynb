```python
# Calculating the average price for each of the 84 property types
average_price_all_property_types = train_df.groupby('property_type')['price'].mean().sort_values()

# Determining thresholds for the three categories for all property types
percentiles_all_types = average_price_all_property_types.quantile([0.33, 0.66])

# Thresholds for all property types
low_price_threshold_all_types = percentiles_all_types.iloc[0]
high_price_threshold_all_types = percentiles_all_types.iloc[1]
def categorize_property_type(price):
    if price <= low_price_threshold_all_types:
        return "Budget-Friendly"
    elif price <= high_price_threshold_all_types:
        return "Mid-Range"
    else:
        return "High-End"
# Categorizing all 84 property types
property_type_categories_all_types = average_price_all_property_types.apply(categorize_property_type)

property_type_categories_all_types.head(20)  # Displaying the first 20 for brevity

# Displaying the categorization for all 84 property types
property_type_categories_full = property_type_categories_all_types.reset_index()
property_type_categories_full.columns = ['Property Type', 'Category']
property_type_categories_full

```

```python
# Splitting the categorized property types into three separate lists
budget_friendly_list = property_type_categories_full[property_type_categories_full['Category'] == 'Budget-Friendly']['Property Type'].tolist()
mid_range_list = property_type_categories_full[property_type_categories_full['Category'] == 'Mid-Range']['Property Type'].tolist()
high_end_list = property_type_categories_full[property_type_categories_full['Category'] == 'High-End']['Property Type'].tolist()

budget_friendly_list, mid_range_list, high_end_list
```

```python
# train set
# turn the value of property_type into budget_friendly if the value is in the budget_friendly_list and the value of property_type into mid_range if the value is in the mid_range_list and  the value of property_type into high_end if the value is in the high_end_list
train_df['property_type'] = train_df['property_type'].apply(lambda x: 'budget_friendly' if x in budget_friendly_list else ('mid_range' if x in mid_range_list else 'high_end'))

```

```python
#draw heatmap to see multi colinearity correlation between features, doing features selection
def draw_heatmap(df,save_name):
    sns.set(style="white")
    corr = df.corr()
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
    #f, ax = plt.subplots(figsize=figsize)
    fig = plt.figure(figsize=(18,10))
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    sns.heatmap(corr, mask=mask, cmap=cmap,linewidths=0.5,cbar_kws={"shrink": .5},vmax=corr[corr != 1.0].max().max());
    plt.savefig("/content/drive/MyDrive/figure_"+save_name)
draw_heatmap(train_df,'heatmap_1.pdf')
```
