# Kaggle Data Analyzing

## Import

```python
import pandas as pd
import numpy as np
import random
import ast
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import MinMaxScaler
```

## Set Random Seed

```python
def set_all_seeds(RANDOM_SEED):
    random.seed(RANDOM_SEED)
    np.random.seed(RANDOM_SEED)
seed = 1
set_all_seeds(seed)
```

## Data Analysis

```python
# get original data
train_data = pd.read_csv("data/train.csv")
train_data.head()
```

```python
# delete obviously useless features
drop_columns = ['id','scrape_id','last_scraped','picture_url','host_id','host_name','name']
train_data = train_data.drop(drop_columns, axis=1)
train_data.head()
```

```python
# check nan
train_data.isnull().sum()
```

```python
# deal with incomplete data
# use the most frequent element to fill the nan
most_frequent_with_nans = ['host_is_superhost', 'bathrooms_text']
imputer = SimpleImputer(strategy='most_frequent')
train_data[most_frequent_with_nans] = imputer.fit_transform(train_data[most_frequent_with_nans])

# use median to fill the nan
beds_imputer = SimpleImputer(strategy='median')
train_data['beds'] = beds_imputer.fit_transform(train_data[['beds']])

# fill nan with empty string, because only the length of description is used
train_data['description'] = train_data['description'].fillna('')

train_data.isnull().sum()
```

```python
# add feature description_length and drop description
# here only the length of description is used, because the house with more detailed description often has higher price
train_data['description_length'] = train_data['description'].apply(len)
train_data = train_data.drop('description', axis=1)
train_data.head()
```

```python
# check the dtypes
train_data.dtypes
```

```python
# deal with host_since
# extract the year and month, ignore day (because I think day has very little impact)
train_data['host_since'] = pd.to_datetime(train_data['host_since'])
train_data['host_since'] = train_data['host_since'].dt.year + train_data['host_since'].dt.month / 12
train_data['host_since'] = train_data['host_since'].astype('float64')
```

```python
# deal with bathrooms_text
# extract number from bathrooms_text
train_data['bathrooms_text'] = train_data['bathrooms_text'].replace(to_replace='half', value='0.5', regex=True)
train_data['bathrooms_text'] = train_data['bathrooms_text'].replace(to_replace='Half', value='0.5', regex=True)
train_data['bathrooms_text'] = train_data['bathrooms_text'].str.extract('(\d+\.?\d*)').astype('float64')
train_data['bathrooms_text'] = train_data['bathrooms_text'].fillna(train_data['bathrooms_text'].mean())
```

```python
# add feature amenities_count
train_data['amenities_count'] = train_data['amenities'].apply(lambda x: len(x.split(',')))
```

```python
# delete property_type and neighbourhood_cleansed, because they have too many values, which is hard to onehot
train_data = train_data.drop(['property_type','neighbourhood_cleansed'], axis=1)
```

```python
# onhot encode all 'object' features (except amenities)
categorical_columns = train_data.select_dtypes(include=['object']).columns
categorical_columns_to_encode = [col for col in categorical_columns if len(train_data[col].unique()) <= 20]
print(categorical_columns_to_encode)
train_data = pd.get_dummies(train_data, columns=categorical_columns_to_encode, drop_first=True)
train_data.head()
```

```python
# GMM
# Cluster the currently processed features and treat the GMM result as a new feature
# normalize
x = train_data.drop(['price', 'amenities'], axis=1)
scaler = MinMaxScaler()
x = scaler.fit_transform(x)

# determine n_components
neg_log_likelihoods = []
for i in range(1, 11):
    gmm = GaussianMixture(n_components=i, covariance_type='full', init_params='kmeans', random_state=seed).fit(x)
    neg_log_likelihoods.append(-gmm.score(x))

# plot K vs Negative Log-likelihood
plt.plot(range(1, 11), neg_log_likelihoods, marker='o')
plt.xticks(range(1, 11))
plt.title('K vs Negative Log-likelihood')
plt.xlabel('K')
plt.ylabel('Negative Log-likelihood')
plt.grid(True)
plt.show()

# K = 6 is a proper choice
gmm = GaussianMixture(n_components=6, covariance_type='full', init_params='kmeans', random_state=seed).fit(x)
train_data['gmm_cluster'] = gmm.predict(x)
train_data.head()
```

```python
# From the figures above, only accommodates, bathrooms_text and beds have positively correlation with price.
# Based on the figures above, features below are dropped.
# After testing, it has been found that removing the following features helps to improve accuracy.
train_data = train_data.drop(['minimum_nights','maximum_nights','minimum_minimum_nights','maximum_minimum_nights',
                              'minimum_maximum_nights','maximum_maximum_nights','availability_60','availability_90',
                              'number_of_reviews_ltm','number_of_reviews_l30d'], axis=1)
```

```python
# chnage amenities to list
train_data['amenities'] = train_data['amenities'].apply(ast.literal_eval)

# extract amenities with a count of more than 1000
count = train_data['amenities'].explode().value_counts()
amenities_cols = []
for i in count.keys():
    if count[i] > 1000:
        amenities_cols.append(i)
print(amenities_cols)
```

```python
import warnings
warnings.filterwarnings('ignore')

# onehot encoding
train_data_amenities_price = pd.DataFrame()
train_data_amenities_price['price'] = train_data['price']
for i in amenities_cols: 
    train_data_amenities_price[i] = 0
for i in train_data.index:
    for item in amenities_cols:
        if item in train_data.loc[i, 'amenities']:
            train_data_amenities_price.loc[i, item] = 1

# get correlation of each feature and price, and sort
index = []
val = []
correlation_with_price = train_data_amenities_price.corrwith(train_data_amenities_price['price']).sort_values(ascending=False)
for i in correlation_with_price.index:
    if correlation_with_price[i] > 0.15 or correlation_with_price[i] < -0.15:
        if i != 'price':
            index.append(i)
            val.append(correlation_with_price[i])
data = {
    'correlation': val
}
correlation_df = pd.DataFrame(data, index=index)
```

```python
# correlations between these selected amenities and price
x = train_data_amenities_price
seaborn.set(style="white")
corr = x.corr()
mask = np.zeros_like(corr, dtype='bool')
fig = plt.figure(figsize=(20,12))
cmap = seaborn.diverging_palette(100, 0, as_cmap=True)
seaborn.heatmap(corr, mask=mask, cmap=cmap, linewidths=0.1, cbar_kws={"shrink": 1})
```

```python
# add amenities features if its correlation with price > 0.15 or < -0.15
amenities_cols = ['Lock on bedroom door', 'Indoor fireplace', 'Dishwasher', 'BBQ grill', 'Barbecue utensils', 'Fire pit',
                  'Outdoor dining area', 'Sun loungers', 'Outdoor furniture', 'Private patio or balcony', 'Private backyard – Fully fenced', 'Pool']

for i in amenities_cols: 
    train_data[i] = 0
for i in train_data.index:
    for item in amenities_cols:
        if item in train_data.loc[i, 'amenities']:
            train_data.loc[i, item] = 1
train_data = train_data.drop(['amenities'], axis=1)
train_data.head()
```
