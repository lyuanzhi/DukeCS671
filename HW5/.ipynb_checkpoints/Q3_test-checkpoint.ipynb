```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

w1, w2 = 500, -500
b1 = -w1 * 0.4
b2 = -w2 * 0.6
w3, w4 = 1, 1
b3 = -1

x = np.linspace(0, 1, 1000)
y = w3 * sigmoid(w1 * x + b1) + w4 * sigmoid(w2 * x + b2) + b3

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Bump Function Approximation')
plt.axvline(0.4, color='red', linestyle='--')
plt.axvline(0.6, color='red', linestyle='--')
plt.title('Neural Network Approximation of Bump Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

w11, w21, w12, w22 = 500, 0, -500, 0
w3, w4 = 1, 1
b1 = -w11 * 0.3
b2 = -w12 * 0.7
b3 = -1

def f(x1, x2):
    return w3 * sigmoid(w11 * x1 + w21 * x2 + b1) + w4 * sigmoid(w12 * x1 + w22 * x2 + b2) + b3

x1 = np.linspace(0, 1, 1000)
x2 = np.linspace(0, 1, 1000)

X1, X2 = np.meshgrid(x1, x2)
Z = np.array([[f(i, j) for i in x1] for j in x2])

# Plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X2, X1, Z, cmap=plt.get_cmap('rainbow'))
fig.colorbar(surf)
ax.set_xlabel('x2')
ax.set_ylabel('x1')
ax.set_zlabel('f(x1,x2)')
ax.set_title('Neural Network Approximation of Bump Function')
ax.view_init(elev=30, azim=-30)
plt.show()
```
